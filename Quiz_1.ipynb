```python
column_to_analyze = df['EdLevel']
```

```python
word_counts = column_to_analyze.value_counts()
print(word_counts)
```

```python
column_for_count = df['Age']
word_count = column_for_count.value_counts()
print(word_count)
```

```python
condition1 = df['EdLevel'] == 'Professional degree (JD, MD, Ph.D, Ed.D, etc.)'
condition2 = df['Age'] == '45-54 years old'

combined_condition = condition1 & condition2

filtered_df = df[combined_condition]

filtered_df
```

```python
sorted_df = filtered_df.sort_values(by=['RemoteWork', 'LearnCode'])
sorted_df
```

```python
column_name = 'RemoteWork'
```

```python
missing_values_count = sorted_df[column_name].isna().sum()
total_values = len(sorted_df[column_name])
```

```python
percentage_missing = (missing_values_count / total_values) * 100

print(f"Percentage of missing values in '{column_name}': {percentage_missing:.2f}%")
```

```python
cleaned_df = sorted_df.dropna(subset=[column_name])
cleaned_df
```

```python
only_columns = ['RemoteWork', 'Employment']
cleaned_df[only_columns]
```

სტატისტიკური ფუნქციები (mean, standard
deviation, median, min, max).

```python
word_counts
```

```python
word_count_df = pd.DataFrame(list(word_counts.items()), columns=['Word', 'Count'])
word_count_df
```

```python
numeric_column_name = 'Count'
```

```python
mean_value = word_count_df[numeric_column_name].mean()
print("mean:", mean_value)
```

```python
std_dev_value = word_count_df[numeric_column_name].std()
print("standard Deviation:", std_dev_value)
```

```python
median_value = word_count_df[numeric_column_name].median()
print("median:", median_value)
```

```python
min_value = word_count_df[numeric_column_name].min()
print("minimum:", min_value)
```

```python
max_value = word_count_df[numeric_column_name].max()
print("maximum:", max_value)
```

```python
import matplotlib.pyplot as plt
import re
```

```python
# Bar chart
plt.figure(figsize=(10, 6))

word_count_df['Word'] = word_count_df['Word'].apply(lambda x: re.sub(r'\([^()]*\)', '', x))

plt.bar(word_count_df['Word'], word_count_df['Count'], color='indigo')
plt.xlabel('Degree')
plt.ylabel('Count')
plt.title('Word Counts')
plt.xticks(rotation=45, ha='right', fontsize=7)
plt.tight_layout()
plt.show()
```

```python
# Line chart
plt.figure(figsize=(10, 6))
plt.plot(word_count_df['Word'], word_count_df['Count'], marker='o', color='green', linestyle='-')
plt.xlabel('Word')
plt.ylabel('Count')
plt.title('Word Counts')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
